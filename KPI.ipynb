DDOG

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ticker_symbol = "DDOG"

stock = yf.Ticker(ticker_symbol)
stock_df = stock.history(start="2025-01-01", end="2025-12-31")
```

```python
stock_df.head()
```

```python
stock_df.info()
```

```python
stock_df.describe()
```

```python
plt.figure(figsize=(10,6))
plt.plot(stock_df.index, stock_df["Close"])
plt.title("DDOG Closing Price")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.grid(True)
plt.show()
```

```python
stock_df.isnull().sum()
```

```python
completeness = (1 - stock_df.isnull().sum().sum() / (stock_df.shape[0] * stock_df.shape[1])) * 100
print("Completeness:", completeness, "%")
```

```python
print("Start date:", stock_df.index.min())
print("End date:", stock_df.index.max())
```

```python
print("Minimum closing price:", stock_df["Close"].min())
print("Maximum closing price:", stock_df["Close"].max())
```

```python
print("Negative values in Close:", (stock_df["Close"] < 0).sum())
```

```python
print("Duplicated rows:", stock_df.duplicated().sum())
print("Is index sorted:", stock_df.index.is_monotonic_increasing)
```

```python
stock_df.to_csv("DDOG_data.csv")
```
